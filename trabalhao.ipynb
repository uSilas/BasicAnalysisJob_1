__1.CRIANDO TABELA__

```python
import pandas as pd
import matplotlib.pyplot as plt

pagina_1 = pd.read_csv('./Introdução a Ciência de Dados - Trabalho 1 - KinderJoy.csv')
pagina_2 = pd.read_csv('./Introdução a Ciência de Dados - Trabalho 1 - KinderBueno.csv')
pagina_3 = pd.read_csv('./Introdução a Ciência de Dados - Trabalho 1 - KitKat.csv')
pagina_1.head()

pagina_1['Preço Original'] = pagina_1['Preço Original'].str.replace('R$', '', regex=False).str.replace(',', '.').astype(float)
pagina_2['Preço Original'] = pagina_2['Preço Original'].str.replace('R$', '', regex=False).str.replace(',', '.').astype(float)
pagina_3['Preço Original'] = pagina_3['Preço Original'].str.replace('R$', '', regex=False).str.replace(',', '.').astype(float)


media = [pagina_1['Preço Original'].mean(),pagina_2['Preço Original'].mean(),pagina_3['Preço Original'].mean()]
mediana = [pagina_1['Preço Original'].median(),pagina_2['Preço Original'].median(),pagina_3['Preço Original'].median()]
maximo = [pagina_1['Preço Original'].max(),pagina_2['Preço Original'].max(),pagina_3['Preço Original'].max()]
minimo = [pagina_1['Preço Original'].min(),pagina_2['Preço Original'].min(),pagina_3['Preço Original'].min()]
diff = []
for i in range(3):
    print(i)
    valor = maximo[i] - minimo[i]
    diff.append(valor)

std = [pagina_1['Preço Original'].std(),pagina_2['Preço Original'].std(),pagina_3['Preço Original'].std()]
print(media,mediana,maximo,minimo,diff,std)
dados = [media,mediana,maximo,minimo,diff,std]
colunas = ['KinderJoy','KinderBueno','KitKat']
linhas = ['Media','Mediana', 'Máximo', 'Mínimo', 'Amplitude','Desvio Padrão']

analise_df = pd.DataFrame(dados,columns=colunas,index=linhas)
analise_df

```

__2.PLOTANDO GRÁFICOS__

```python
plt.figure(figsize=(15, 10))

estatisticas = analise_df.index

for i, estatistica in enumerate(estatisticas):
    plt.subplot(2, 3, i + 1) 
    plt.bar(analise_df.columns, analise_df.loc[estatistica], color=['skyblue', 'lightgreen', 'lightcoral'])
    plt.title(estatistica.capitalize())
    plt.ylabel('Valor')
    plt.xticks(rotation=45)

plt.tight_layout()


plt.show()
```
